## Plotting the inequality

This plots the inequality

$$
    (1 - 2p)^2 \ge \frac{m}{m + 1.96^2}
$$ 

which is the chance of a circuit being satisfiable if Deutsch-Jozsa does not consistently return a state with at least one |1>.

```python
import numpy as np
import matplotlib.pyplot as plt

m = np.linspace(0, 1000, num=1000)
p = lambda m: (np.sqrt(m/(m + (1.96 ** 2))) - 1) / -2

plt.figure(dpi=400)
plt.xlabel("m")
plt.ylim([0, 0.08])
plt.xlim([0, 500])
plt.ylabel("p")
plt.plot(p(m))
```
